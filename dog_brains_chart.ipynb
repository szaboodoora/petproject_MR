```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set(style="ticks", color_codes=True)
```

```python
df=pd.read_excel("C:/Users/dooora/Desktop/jupyter/Kalman-brain_chart.xlsx",index_col='ID')
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'Volume\n(cm3)': 'Volume','Surface\n(cm2)': 'Surface', 'skull index': 'skull_index','skull length [mm]': 'skull_length', 'skull width [mm]': 'skull_width'}, inplace=True)
```

```python
ax = sns.scatterplot(x="Surface", y="Volume", legend=False, data=df)


plt.title('Measured subjects')
# Set x-axis label
plt.xlabel('brain surface [cm2]')
# Set y-axis label
plt.ylabel('brain volume [cm3]')
```

```python
ax = sns.scatterplot(x="Surface", y="Volume", hue="breed", s=300, data=df)
plt.legend(loc='upper left')

plt.title('Measured subjects')
# Set x-axis label
plt.xlabel('brain surface [cm2]')
# Set y-axis label
plt.ylabel('brain volume [cm3]')
```

```python
ax = sns.scatterplot(x="Surface", y="Volume", hue="breed", s=300, legend=False, data=df)
sns.set(rc={'figure.figsize':(15,15)}, font_scale=2)
sns.set_style( {'font.family': ['Calibri']})

for line in range(0,df.shape[0]):
     ax.text(df.Surface[line]+0.01, df.Volume[line]-1, 
     df.breed[line], horizontalalignment='left', 
     size='smaller', color='black')
        
plt.title('Measured subjects')
# Set x-axis label
plt.xlabel('brain surface [cm2]')
# Set y-axis label
plt.ylabel('brain volume [cm3]')

```
